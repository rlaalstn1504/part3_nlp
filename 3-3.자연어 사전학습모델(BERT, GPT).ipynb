#### ëª¨ë¸ê³¼ í† í¬ë‚˜ì´ì € ë¶ˆëŸ¬ì˜¤ê¸°

Hugging Faceì˜ transformers ë¼ì´ë¸ŒëŸ¬ë¦¬ì—ì„œ ì‚¬ì „í•™ìŠµëœ ëª¨ë¸(AutoModel)ê³¼ í† í¬ë‚˜ì´ì €(AutoTokenizer)ë¥¼ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤.

````python
import torch
# ì‚¬ì „í•™ìŠµëœ(pretrained) BERT ëª¨ë¸ê³¼ í† í¬ë‚˜ì´ì €ë¥¼ ì‰½ê²Œ ë¶ˆëŸ¬ì˜¤ëŠ” í´ë˜ìŠ¤ë“¤ì„ ê°€ì ¸ì˜µë‹ˆë‹¤.
from transformers import AutoModelForSequenceClassification, AutoTokenizer

# ì‚¬ìš©í•  ì‚¬ì „í•™ìŠµ ëª¨ë¸ ì´ë¦„ ì§€ì •
# "bert-base-uncased"ëŠ” ì†Œë¬¸ì ê¸°ë°˜ ì˜ì–´ BERT ëª¨ë¸ (12-layer, 110M parameters)
model_name = "bert-base-uncased"

# ì‚¬ì „í•™ìŠµëœ í† í¬ë‚˜ì´ì € ë¶ˆëŸ¬ì˜¤ê¸°
# â†’ ì…ë ¥ ë¬¸ì¥ì„ WordPiece ë‹¨ìœ„ë¡œ ìª¼ê°œì–´ ìˆ«ì ID ì‹œí€€ìŠ¤ë¡œ ë³€í™˜í•¨
tokenizer = AutoTokenizer.from_pretrained(model_name)

# ì‚¬ì „í•™ìŠµëœ ëª¨ë¸ ë¶ˆëŸ¬ì˜¤ê¸°
# â†’ ë¬¸ì¥ ë¶„ë¥˜(Sequence Classification)ìš© BERT êµ¬ì¡° ë¡œë“œ
# from_pretrained()ëŠ” ë‚´ë¶€ì ìœ¼ë¡œ Hugging Face Hubì—ì„œ ëª¨ë¸ ê°€ì¤‘ì¹˜ë¥¼ ë‹¤ìš´ë¡œë“œí•¨
model = AutoModelForSequenceClassification.from_pretrained(model_name)
````

````python
# ì˜ˆì‹œ ë¬¸ì¥ í•˜ë‚˜ ì¤€ë¹„
text = "Hugging Face is creating a great ecosystem for NLP."

#  í† í¬ë‚˜ì´ì €ë¡œ í…ìŠ¤íŠ¸ ì „ì²˜ë¦¬
#  - ë¬¸ì¥ì„ BERTê°€ ì´í•´í•  ìˆ˜ ìˆëŠ” ìˆ«ì í† í°(id)ìœ¼ë¡œ ë³€í™˜
#  - return_tensors="pt" â†’ PyTorch í…ì„œ(torch.tensor) í˜•íƒœë¡œ ë°˜í™˜
#  - ê²°ê³¼: {'input_ids': tensor(...), 'attention_mask': tensor(...)} í˜•íƒœì˜ ë”•ì…”ë„ˆë¦¬
inputs = tokenizer(text, return_tensors="pt")
````

````python
# ëª¨ë¸ì— ì…ë ¥ ì „ë‹¬ (ì¶”ë¡  ë‹¨ê³„)
#  - **inputs ë”•ì…”ë„ˆë¦¬ì˜ í‚¤**(input_ids, attention_mask ë“±)ë¥¼ ê·¸ëŒ€ë¡œ í’€ì–´ì„œ ì „ë‹¬(**inputs)
#  - ëª¨ë¸ì€ ì…ë ¥ ë¬¸ì¥ì„ ì¸ì½”ë”©í•˜ì—¬ ê° í´ë˜ìŠ¤(ì˜ˆ: ê¸ì •/ë¶€ì •)ì— ëŒ€í•œ ì ìˆ˜ë¥¼ ê³„ì‚°
outputs = model(**inputs)
````

````python
# ì¶œë ¥ ê²°ê³¼ í™•ì¸
#  - outputsëŠ” transformersì˜ SequenceClassifierOutput ê°ì²´
#  - outputs.logitsëŠ” (ë°°ì¹˜ í¬ê¸°, í´ë˜ìŠ¤ ê°œìˆ˜) í˜•íƒœì˜ ì ìˆ˜(logit) í…ì„œ
#  - ì˜ˆ: tensor([[-1.203, 2.315]]) â†’ ë‘ í´ë˜ìŠ¤(NEGATIVE, POSITIVE)ì— ëŒ€í•œ ì ìˆ˜
#  - softmaxë¥¼ ì ìš©í•˜ë©´ ê° í´ë˜ìŠ¤ì˜ í™•ë¥ ë¡œ ë³€í™˜í•  ìˆ˜ ìˆìŒ
print(outputs.logits)
````

````python
# pipeline()ì„ ì‚¬ìš©í•˜ë©´, ë³µì¡í•œ ì „ì²˜ë¦¬ë‚˜ ëª¨ë¸ ë¡œë“œë¥¼ ëª°ë¼ë„ ë‹¨ í•œ ì¤„ë¡œ ë‹¤ì–‘í•œ NLP ì‘ì—…ì„ ìˆ˜í–‰í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

from transformers import pipeline

# ê°ì •ë¶„ì„ ì˜ˆì‹œ
classifier = pipeline("sentiment-analysis")
result = classifier("I love using Hugging Face models!")
print(result)

# ì¶œë ¥ ì˜ˆì‹œ:
#[{'label': 'POSITIVE', 'score': 0.9998}] 
````

#### Fine-tuningì„ ìœ„í•œ ê¸°ë³¸ êµ¬ì¡°

íŒŒì¸íŠœë‹ ì‹œ ë§ì´ ì‚¬ìš©í•˜ëŠ” êµ¬ì„±ì€ **Trainer API**ì…ë‹ˆë‹¤.

````python
# transformers ë¼ì´ë¸ŒëŸ¬ë¦¬ì—ì„œ íŒŒì¸íŠœë‹(Fine-tuning)ì„ ìœ„í•œ ì£¼ìš” í´ë˜ìŠ¤ ë¶ˆëŸ¬ì˜¤ê¸°
#  - AutoModelForSequenceClassification : ë¬¸ì¥ ë¶„ë¥˜ìš© ì‚¬ì „í•™ìŠµ ëª¨ë¸
#  - Trainer / TrainingArguments : í•™ìŠµ ê³¼ì •ì„ ë‹¨ìˆœí™”ì‹œì¼œì£¼ëŠ” ê³ ìˆ˜ì¤€ API
#  - load_dataset : ê³µê°œ ë°ì´í„°ì…‹ì„ ì‰½ê²Œ ë¶ˆëŸ¬ì˜¬ ìˆ˜ ìˆëŠ” í•¨ìˆ˜ (datasets ë¼ì´ë¸ŒëŸ¬ë¦¬)
from transformers import AutoModelForSequenceClassification, AutoTokenizer, Trainer, TrainingArguments
from datasets import load_dataset


# ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸° -------------------------------------------------------------
# GLUE(SST-2) ë°ì´í„°ì…‹ ë¡œë“œ
#  - GLUE : General Language Understanding Evaluation (ì–¸ì–´ ì´í•´ ë²¤ì¹˜ë§ˆí¬)
#  - SST-2 : Stanford Sentiment Treebank v2 (ê°ì •ë¶„ì„ ë°ì´í„°ì…‹)
#  - ì…ë ¥: ë¬¸ì¥(sentence)
#  - ì¶œë ¥: ê°ì • ë¼ë²¨(label) â†’ 0(ë¶€ì •) / 1(ê¸ì •)
dataset = load_dataset("glue", "sst2")

# dataset êµ¬ì¡° ì˜ˆì‹œ:
# dataset['train'][0]  â†’ {'sentence': 'a stirring , funny and finally transporting re-imagining of beauty and the beast and 1930s horror films', 'label': 1}
````

````python
# ëª¨ë¸ ë° í† í¬ë‚˜ì´ì € ë¶ˆëŸ¬ì˜¤ê¸° -----------------------------------------------
model_name = "bert-base-uncased"   # ì˜ì–´ ì†Œë¬¸ì ê¸°ë°˜ BERT ëª¨ë¸ (12-layer, 110M parameters)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# AutoModelForSequenceClassification : ë¬¸ì¥ ë¶„ë¥˜ íƒœìŠ¤í¬ìš© BERT
# num_labels=2 â†’ ê¸ì •/ë¶€ì • ë‘ ê°€ì§€ ê°ì • í´ë˜ìŠ¤ë¥¼ ì˜ˆì¸¡í•˜ë„ë¡ ì¶œë ¥ì¸µ(Linear(768â†’2)) ì¶”ê°€
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)
````

````python
# ë°ì´í„° ì „ì²˜ë¦¬ (í† í¬ë‚˜ì´ì§•)
# BERTëŠ” ë¬¸ì¥ì„ ë‹¨ì–´ê°€ ì•„ë‹Œ WordPiece ë‹¨ìœ„ë¡œ ìª¼ê°œì„œ ì…ë ¥ë°›ìŒ
# truncation=True â†’ ë¬¸ì¥ì´ ë„ˆë¬´ ê¸¸ë©´ ìµœëŒ€ ê¸¸ì´(512í† í°)ê¹Œì§€ë§Œ ìë¦„
# padding=True â†’ ë°°ì¹˜(batch) ë‚´ ë¬¸ì¥ ê¸¸ì´ë¥¼ ë™ì¼í•˜ê²Œ ë§ì¶¤
def preprocess_function(examples):
    return tokenizer(examples["sentence"], truncation=True, padding=True)

# dataset.map()ì„ í†µí•´ ì „ì²´ ë°ì´í„°ì…‹ì— ì „ì²˜ë¦¬ í•¨ìˆ˜ë¥¼ ì ìš©
# batched=True â†’ ì—¬ëŸ¬ ìƒ˜í”Œì„ í•œ ë²ˆì— ì²˜ë¦¬í•˜ì—¬ ì†ë„ í–¥ìƒ
encoded_dataset = dataset.map(preprocess_function, batched=True)

# encoded_dataset ì˜ˆì‹œ:
# {'input_ids': [...], 'attention_mask': [...], 'label': ...}


# í•™ìŠµ ì„¤ì • (TrainingArguments)
training_args = TrainingArguments(
    output_dir="./results",          # í•™ìŠµ ê²°ê³¼(ëª¨ë¸, ë¡œê·¸) ì €ì¥ ê²½ë¡œ
    eval_strategy="epoch",     # í•œ epochë§ˆë‹¤ ê²€ì¦(validation) ìˆ˜í–‰
    learning_rate=2e-5,              # í•™ìŠµë¥  (BERT ê¶Œì¥ê°’: 2e-5 ~ 5e-5)
    per_device_train_batch_size=32,  # GPUë‹¹ ë°°ì¹˜ í¬ê¸°
    num_train_epochs=2,              # ì „ì²´ ë°ì´í„°ì…‹ ë°˜ë³µ í•™ìŠµ íšŸìˆ˜
    weight_decay=0.01,               # L2 ì •ê·œí™”(ê°€ì¤‘ì¹˜ ê°ì‡ ) ê³„ìˆ˜ â†’ ê³¼ì í•© ë°©ì§€
)
from transformers import DataCollatorWithPadding

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

# Trainer ê°ì²´ ìƒì„± ë° í•™ìŠµ ì‹œì‘ 
# TrainerëŠ” í•™ìŠµ ë£¨í”„, ì†ì‹¤ ê³„ì‚°, ê²€ì¦, ë¡œê¹… ë“±ì„ ìë™ìœ¼ë¡œ ì²˜ë¦¬í•˜ëŠ” high-level API
trainer = Trainer(
    model=model,                           # í•™ìŠµí•  ëª¨ë¸
    args=training_args,                    # í•™ìŠµ ì„¤ì •
    train_dataset=encoded_dataset["train"],# í•™ìŠµ ë°ì´í„°ì…‹
    eval_dataset=encoded_dataset["validation"], # ê²€ì¦ ë°ì´í„°ì…‹
    data_collator=data_collator
)

# í•™ìŠµ ì‹œì‘ (fine-tuning)
# ë‚´ë¶€ì ìœ¼ë¡œ optimizer, scheduler, evaluation loopê¹Œì§€ ìë™ ì‹¤í–‰
trainer.train()
````

````python
# í‰ê°€(Evaluation)
# í•™ìŠµëœ ëª¨ë¸ì˜ ì„±ëŠ¥ì„ ê²€ì¦(validation set) ë°ì´í„°ë¡œ í‰ê°€
# Trainer.evaluate()ëŠ” ìë™ìœ¼ë¡œ ì†ì‹¤(loss) ë° ì •í™•ë„(metrics)ë¥¼ ê³„ì‚°
# (metricsë¥¼ ë³„ë„ë¡œ ì„¤ì •í•˜ì§€ ì•Šìœ¼ë©´ lossë§Œ ì¶œë ¥ë¨)

eval_results = trainer.evaluate()
print(" í‰ê°€ ê²°ê³¼:")
print(eval_results)
# ì¶œë ¥ ì˜ˆì‹œ:
# {'eval_loss': 0.35, 'eval_runtime': 15.2, 'eval_samples_per_second': 200.5, ...}
````

````python
# ìƒˆë¡œìš´ ë¬¸ì¥ì— ëŒ€í•œ ì¶”ë¡ (Inference)
# í•™ìŠµëœ ëª¨ë¸ì„ ì´ìš©í•´ ê°ì •(ê¸ì •/ë¶€ì •)ì„ ì˜ˆì¸¡

text = "The movie was surprisingly good and emotional."

# ì…ë ¥ ë¬¸ì¥ í† í°í™” (ëª¨ë¸ ì…ë ¥ í˜•íƒœë¡œ ë³€í™˜)
inputs = tokenizer(text, return_tensors="pt")

# ëª¨ë¸ì„ í‰ê°€ ëª¨ë“œë¡œ ë³€ê²½ (Dropout ë“± ë¹„í™œì„±í™”)
model.eval()

# ì¶”ë¡  ì‹¤í–‰ (Gradient ê³„ì‚° ë¹„í™œì„±í™” â†’ ë” ë¹ ë¥´ê³  ë©”ëª¨ë¦¬ ì ˆì•½)
with torch.no_grad():
    outputs = model(**inputs)

# ëª¨ë¸ ì¶œë ¥(logits)ì€ ê° í´ë˜ìŠ¤(0=ë¶€ì •, 1=ê¸ì •)ì— ëŒ€í•œ ì ìˆ˜ê°’
logits = outputs.logits

# Softmaxë¡œ í™•ë¥ í™” í›„ argmaxë¡œ ì˜ˆì¸¡ í´ë˜ìŠ¤ ê²°ì •
predicted_class = torch.argmax(torch.softmax(logits, dim=1)).item()

# ê²°ê³¼ ì¶œë ¥
labels = ["Negative", "Positive"]
print(f"\nğŸ“ ì…ë ¥ ë¬¸ì¥: {text}")
print(f"ğŸ”® ì˜ˆì¸¡ ê²°ê³¼: {labels[predicted_class]} (í™•ë¥ : {torch.softmax(logits, dim=1).max().item():.4f})")
````

from transformers import GPT2Tokenizer, GPT2LMHeadModel

# ë²ˆì—­


````python
#  GPT2 ê¸°ë°˜ í•œ-ì˜ ë²ˆì—­ íŒŒì¸íŠœë‹ ì „ì²´ ì½”ë“œ (ì™„ì „ ì •ë¦¬ë³¸)

import torch
from datasets import load_dataset
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    Trainer,
    TrainingArguments,
)

device = "cuda" if torch.cuda.is_available() else "cpu"
print("DEVICE =", device)

# 1. GPT-2 ëª¨ë¸ / í† í¬ë‚˜ì´ì € ë¡œë“œ + pad/eos ì²˜ë¦¬
base_model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(base_model_name)

# gpt2ëŠ” pad_tokenì´ ì—†ìŒ â†’ eos_tokenì„ ëŒ€ì‹  padë¡œ ì„¤ì •
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
    tokenizer.pad_token_id = tokenizer.eos_token_id

model = AutoModelForCausalLM.from_pretrained(base_model_name).to(device)

# 2. ê°„ë‹¨ ë²ˆì—­ í•¨ìˆ˜ (í›ˆë ¨ ì „/í›„ ë¹„êµìš©)

def gpt_translate(model, tokenizer, text_ko: str, max_new_tokens: int = 64):
    prompt = (
        "Translate the following Korean sentence into natural English.\n"
        f"Korean: {text_ko}\n"
        "English:"
    )

    input_ids = tokenizer(
        prompt,
        return_tensors="pt",
        truncation=True,
        max_length=256,
    ).input_ids.to(model.device)

    with torch.no_grad():
        output_ids = model.generate(
            input_ids,
            max_new_tokens=max_new_tokens,
            do_sample=True,
            top_p=0.9,
            temperature=0.7,
            pad_token_id=tokenizer.pad_token_id,
        )

    generated = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    return generated.split("English:")[-1].strip()


# 3. ë°ì´í„°ì…‹ ë¡œë“œ
dataset = load_dataset("strongminsu/ko-en-structured-translations")

SRC_COL = "ko"
TGT_COL = "en"

train_dataset = dataset["train"]
valid_dataset = dataset["test"]   # ê²€ì¦ì…‹ ëŒ€ì²´
test_dataset  = dataset["validation"]


# 4. íŒŒì¸íŠœë‹ ì „ ë²ˆì—­ ê²°ê³¼ í™•ì¸
sample_texts = [
    "ì´ ëª¨ë¸ì€ í•œêµ­ì–´ë¥¼ ì˜ì–´ë¡œ ë²ˆì—­í•˜ëŠ” ë° ì‚¬ìš©ë©ë‹ˆë‹¤.",
    "ì˜¤ëŠ˜ ë‚ ì”¨ê°€ ì •ë§ ì¢‹ë„¤ìš”.",
    "ë‚´ì¼ íšŒì˜ ì¼ì • ë‹¤ì‹œ í•œë²ˆ í™•ì¸í•´ ì£¼ì„¸ìš”.",
]

print("=== íŒŒì¸íŠœë‹ ì „ ë²ˆì—­ ì˜ˆì‹œ ===")
for text in sample_texts:
    print(f"[KO] {text}")
    print(f"[EN] {gpt_translate(model, tokenizer, text)}\n")


# 5. ì „ì²˜ë¦¬ í•¨ìˆ˜ ì •ì˜
#    - prompt + target + eos_token
#    - input_ids / labels / attention_mask ë°˜í™˜
max_length = 256

def preprocess_function(example):
    src = example[SRC_COL]
    tgt = example[TGT_COL]

    # â˜… ko/en ì´ ë¦¬ìŠ¤íŠ¸(list[str])ë¡œ ë“¤ì–´ì˜¤ëŠ” ê²½ìš° ì²˜ë¦¬ (ë¬¸ì œì˜ í•µì‹¬)
    if isinstance(src, list):
        src = " ".join(src)
    if isinstance(tgt, list):
        tgt = " ".join(tgt)

    prompt = (
        "Translate the following Korean sentence into natural English.\n"
        f"Korean: {src}\n"
        "English:"
    )

    text = prompt + " " + tgt + tokenizer.eos_token

    encoded = tokenizer(
        text,
        truncation=True,
        max_length=max_length,
        padding=False,
    )

    ids = encoded["input_ids"]

    return {
        "input_ids": ids,
        "labels": ids.copy(),
        "attention_mask": [1] * len(ids),
    }


# 6. ì „ì²˜ë¦¬ ë°ì´í„°ì…‹ ìƒì„± (ì¤‘ìš”!)
tokenized_train = train_dataset.map(
    preprocess_function,
    remove_columns=train_dataset.column_names
)

tokenized_valid = valid_dataset.map(
    preprocess_function,
    remove_columns=valid_dataset.column_names
)


# 7. Collator (paddingë§Œ ìˆ˜í–‰)
def gpt_collator(features):
    # pad input_ids
    batch = tokenizer.pad(
        {"input_ids": [f["input_ids"] for f in features]},
        padding=True,
        return_tensors="pt"
    )
    
    # pad labels (ì§ì ‘ pad ì²˜ë¦¬)
    max_len = batch["input_ids"].shape[1]
    labels = []
    for f in features:
        pad_len = max_len - len(f["labels"])
        labels.append(f["labels"] + [-100] * pad_len)

    batch["labels"] = torch.tensor(labels)
    return batch


# 8. Trainer ì„¤ì •
output_dir = "./gpt2-ko-en-finetuned"

training_args = TrainingArguments(
    output_dir=output_dir,
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    learning_rate=5e-5,
    num_train_epochs=10,
    eval_strategy="epoch",
    logging_steps=50,
    save_strategy="epoch",
    fp16=torch.cuda.is_available(),
    report_to="none",
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_valid,
    data_collator=gpt_collator,
)


# 9. í•™ìŠµ ì‹œì‘
trainer.train()


# 10. íŒŒì¸íŠœë‹ í›„ ë²ˆì—­ í™•ì¸
print("\n=== íŒŒì¸íŠœë‹ í›„ ë²ˆì—­ ì˜ˆì‹œ ===")
for text in sample_texts:
    print(f"[KO] {text}")
    print(f"[EN] {gpt_translate(model, tokenizer, text)}\n")
````
